```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "Untitled form.csv"
df = pd.read_csv(data, header = None)
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
headers = ["Time","Email","Degree","Employement-Status","Acceptability Above 60","Acceptability b/w 45-60","Acceptability b/w 30-45","Acceptability b/w 15-30","Acceptability Below 15","Will chatgpt replace-work?","Is it realiable","Scope in IT","Scope in Medical","Scope in Robotics","Scope in Military","Scope in Agriculture","Rating out of 5"]
df.columns = headers
df.head(5)
```

```python
df.drop('Time',axis=1, inplace=True)
```

```python
df.head(10)
```

```python
df.describe(include="all")
```

```python
df['Rating out of 5'] = df['Rating out of 5'].astype(int)
df['Acceptability Above 60'] = df['Acceptability Above 60'].astype(int)
df['Acceptability b/w 45-60'] = df['Acceptability b/w 45-60'].astype(int)
df['Acceptability b/w 30-45'] = df['Acceptability b/w 30-45'].astype(int)
df['Acceptability b/w 15-30'] = df['Acceptability b/w 15-30'].astype(int)
df['Acceptability Below 15'] = df['Acceptability Below 15'].astype(int)
df['Will chatgpt replace-work?'] = df['Will chatgpt replace-work?'].astype(int)
df['Is it realiable'] = df['Is it realiable'].astype(int)
df['Scope in IT']=df['Scope in IT'].astype(int)
df['Scope in Medical']=df['Scope in Medical'].astype(int)
df['Scope in Robotics']=df['Scope in Robotics'].astype(int)
df['Scope in Agriculture']=df['Scope in Agriculture'].astype(int)
df['Scope in Military']=df['Scope in Military'].astype(int)
df
```

```python
print("Size/Shape of the dataset: ",df.shape)
print("Checking for null values:\n",df.isnull().sum())
print("Checking Data-type of each column:\n",df.dtypes)
```

```python
df.head()
```

```python
rating = df['Acceptability Above 60']

plt.xlabel('Rating')
plt.ylabel('Number of People')

plt.title('Acceptability above 60')

plt.hist(rating, bins=9)
# Show the plot
plt.show()
```

```python
status_stats = df.groupby('Degree')['Rating out of 5'].agg(['mean', 'count'])

status_values = df['Degree'].unique()

status_table = pd.DataFrame({'Degree': status_values})

status_table = pd.merge(status_table, status_stats, on='Degree')

print(status_table)
```
